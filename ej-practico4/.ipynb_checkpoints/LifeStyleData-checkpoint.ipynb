```python
import pandas as pd

df = pd.read_csv("lifestyle_data.csv")

df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Dimensiones del dataset:", df.shape)
print("\nInformación general del dataset:")
df.info()

print("\nEstadísticas descriptivas:")
display(df.describe(include='all').T)
```

```python
df.hist(figsize=(18, 14), bins=30, color='skyblue', edgecolor='black')
plt.suptitle('Distribución de variables numéricas', fontsize=16)
plt.show()

# === Matriz de correlación ===
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm', annot=False)
plt.title("Matriz de correlación")
plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns
outlier_report = []

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_percentage = len(outliers) / len(df) * 100
    outlier_report.append([col, round(outlier_percentage, 2)])

outlier_df = pd.DataFrame(outlier_report, columns=["Columna", "% de Outliers"])
display(outlier_df.sort_values("% de Outliers", ascending=False))
```

```python
missing_pct = df.isnull().mean() * 100
print("Porcentaje de valores faltantes por columna:")
display(missing_pct.sort_values(ascending=False))

cols_high_nulls = missing_pct[missing_pct > 10].index.tolist()
print("\nColumnas con más del 10% de valores perdidos:", cols_high_nulls)

for col in df.columns:
    if df[col].isna().sum() > 0:
        if df[col].dtype == 'object':
            # Imputación con la moda (valor más frecuente)
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            # Imputación con la mediana (más robusta frente a outliers)
            df[col].fillna(df[col].median(), inplace=True)

print("\nTotal de valores nulos después de la imputación:", df.isna().sum().sum())
```

```python
from scipy.stats.mstats import winsorize

# Seleccionar solo columnas numéricas
num_cols = df.select_dtypes(include='number').columns

# Aplicar winsorize a cada columna (recorta el 5% inferior y superior)
for col in num_cols:
    df[col] = winsorize(df[col], limits=[0.05, 0.05])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))
sns.boxplot(data=df[num_cols])
plt.title("Distribución de variables después de aplicar Winsorize - LifeStyleData")
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd

df_enc = df.copy()

# 1) Target (ordinal)
target_col = "Burns_Calories_Bin"
ord_map = {"Low": 0, "Medium": 1, "High": 2, "Very High": 3}
df_enc[target_col] = df_enc[target_col].map(ord_map)

# 2) Detectar categóricas (object) EXCLUYENDO el target
cat_cols = df_enc.select_dtypes(include="object").columns.tolist()
if target_col in cat_cols:
    cat_cols.remove(target_col)

# 3) Separar por cardinalidad
low_card, high_card = [], []
for c in cat_cols:
    (low_card if df_enc[c].nunique(dropna=True) <= 10 else high_card).append(c)

print("One-Hot (≤10 categorías):", low_card)
print("Frecuencia (>10 categorías):", high_card)

# 4) One-Hot Encoding
df_enc = pd.get_dummies(df_enc, columns=low_card, drop_first=True)

# 5) Codificación por frecuencia para alta cardinalidad
for c in high_card:
    freq = df_enc[c].value_counts(normalize=True)
    df_enc[c] = df_enc[c].map(freq)

print("Shape después de encoding:", df_enc.shape)
df_enc.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Creamos una copia para no alterar el DataFrame original
df_norm = df_enc.copy()

# Definimos el escalador Min-Max
scaler = MinMaxScaler()

# Excluimos la columna objetivo si existe
target_col = "Burns_Calories_Bin"
cols_to_scale = df_norm.columns.drop(target_col)

# Aplicamos la normalización (valores entre 0 y 1)
df_norm[cols_to_scale] = scaler.fit_transform(df_norm[cols_to_scale])

# Verificamos los rangos de los datos
print("Rango después de la normalización (LifeStyleData):")
print(df_norm[cols_to_scale].describe().loc[["min", "max"]])

# Mostramos las primeras filas
df_norm.head()

# Guardamos el dataset procesado
df_norm.to_csv("lifestyle_data_normalized.csv", index=False)
```
